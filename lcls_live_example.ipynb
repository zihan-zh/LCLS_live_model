## Follow Colocho's method for fitting the twiss parameters

```python
def tc(cmd, verbose: bool = False):
    result = tao.cmd(cmd)
    if verbose:
        for l in result:
            print(l)
```

```python
for to in tao.cmd('show top10'):
    print(to)
```

```python
match_here = {
    'BEGBC2': use_BC2,
    'WS28144': use_LI28,
    'WS32': use_LTUH_M
}

match_locs = {}
for mln in match_here.keys():
    match_locs[mln] = \
        float([t for t in tao.cmd('show element ' + mln) if t.startswith('S_start, S:')][0].split(' ')[-1])
```

```python
# match_final_cmd has a run in it!  It DOES do a fit.

do_these_fits = list(match_here.keys()) + list(match_here.keys())
fits = []
print('Fitting...')
for i, fit in enumerate(do_these_fits):
    [tc(cmd) for cmd in match_init_cmds + match_here[fit] + match_final_cmds]
    fits.append(
        (fit, {k:tao.lat_list('*', k) for k in outkeys})
    )
    print('\tFinished {:d}/{:d}'.format(i + 1, len(do_these_fits)))
print(len(fits))
```

```python
def plot_fit_twiss(number, fit_output: dict, name: str, axis: plt.Axes):
    axis.plot(fit_output['ele.s'], fit_output['ele.a.beta'], label = r'$\beta_a$', color='C0')
    axis.plot(fit_output['ele.s'], fit_output['ele.b.beta'], label = r'$\beta_b$', color='C1')
    axis.legend(loc='upper left')
    # Add energy to the rhs
    axt = axis.twinx()
    axt.plot(fit_output['ele.s'], fit_output['ele.e_tot']/1e9, color='magenta')
    if number > 2:
        axt.set_ylabel('Energy (GeV)')
    if number in [2, 5]:
        axis.set_xlabel('s (m)')
    if number < 3:
        axis.set_ylabel('Twiss Beta (m)')
    #itime = isotime()
    efinal = fit_output['ele.e_tot'][-1]/1e9
    # axis.set_title(f'{title} | Time: {ISOTIME} | Final energy: {efinal:.2f} GeV')
    axis.set_title(f'{number + 1} - {name}')
    ax.axvline(match_locs[name], linestyle='dashed', color='k')

def plot_design_twiss(design_fit: dict, axis: plt.Axes):
    axis.plot(design_fit['ele.s'], design_fit['ele.a.beta'], label = r'$\beta_a$', color='C0', linestyle='dashed')
    axis.plot(design_fit['ele.s'], design_fit['ele.b.beta'], label = r'$\beta_b$', color='C1', linestyle='dashed')
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 8), sharex='all', sharey='all')
axs = axs.T.flatten()
for i, (name, fit_out) in enumerate(fits):
    ax = axs[i]
    plot_fit_twiss(i, fit_out, name, ax)
    plot_design_twiss(design_output, ax)
efinal = fit_out['ele.e_tot'][-1]/1e9
fig.suptitle(f'Time: {ISOTIME} | Final energy: {efinal:.2f} GeV', y=0.93)
plt.show()
```
